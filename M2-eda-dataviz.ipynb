### TASK 1: Visualize the relationship between Flight Number and Launch Site


Use the function <code>catplot</code> to plot <code>FlightNumber</code> vs <code>LaunchSite</code>, set the  parameter <code>x</code>  parameter to <code>FlightNumber</code>,set the  <code>y</code> to <code>Launch Site</code> and set the parameter <code>hue</code> to <code>'class'</code>


```python
# Plot a scatter point chart with x axis to be Flight Number and y axis to be the launch site, and hue to be the class value
sns.catplot(y="LaunchSite", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Launch Site",fontsize=20)
plt.title('The Flight Number vs. Launch Site', fontsize = 20)
plt.show()
```

Now try to explain the patterns you found in the Flight Number vs. Launch Site scatter point plots.


We observe that as the number of flights increases, the probability of a successful first stage landing also increases. The launch location is a significant factor; launches from KSC LC 39A have the highest percentage of successful first stage landings, followed closely by VAFB SLC 4E. In contrast, Falcon 9 rockets with fewer than 20 launches, particularly those launched from CCAFS SLC 40, have a higher probability of losing the first stage during the landing process.

```python
def percent(df, col1, col2, val):
    per = (df.loc[df[col1] == val, [col2]].sum()/(df[col1] == val).sum())
    return per
```

```python
print ('Number of successful first stage landing from various Launch Sites')
for item in df['LaunchSite'].unique():
    per = percent(df, 'LaunchSite', 'Class', item).values[0] * 100
    print(f"{item}: {per:.2f}%")
```

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column
orbit_success_rate = df.groupby('Orbit')[['Class']].mean()
orbit_success_rate = orbit_success_rate.sort_values(by='Class', ascending=False).reset_index()

plt.figure(figsize=(8, 5))
ax = sns.barplot(data=orbit_success_rate, x="Orbit", y="Class", hue='Class', palette="viridis")
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', label_type='edge', padding=3)

plt.xlabel("Orbit", fontsize=15)
plt.ylabel("Success Rate", fontsize=15)
plt.title("Success Rate by Orbit", fontsize=18)
plt.xticks(rotation=45)
plt.show()
```

Analyze the ploted bar chart try to find which orbits have high sucess rate.


ES-L1, GEO, HEO, and SSO have the highest success rate at 100%. 
rbit SO has the lowest success rate at 0%, with 1 flight that did not end in a successful first stage landing. 
The majority of rockets were launched into GTO orbit (27 launches), with approximately half (52%) achieving a successful first stage landing. 
For ISS launches (21 launches), 62% resulted in a successful first stage landing. 
VLEO launches (14 launches) had an 86% success rate for first stage landings.
PO (9 launches) and MEO (3 launches) both had a 67% success rate for first stage landings. 
Of the 14 LEO launches, 71% resulted in a successful first stage landing.

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend. 


The function will help you get the year from the date:


```python
# A function to Extract years from the date 
year=[]
def Extract_year(date):
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
df['Year'] = Extract_year(df['Date'])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating line plot with markers
plt.figure(figsize=(10, 4))
lineplot = sns.lineplot(y='Class', x='Year', data=df, marker='o')

# Adding average success rate markers
year = df['Year'].unique()
ydot = df.groupby(by= 'Year')['Class'].mean()
for year, ydot in zip(year, ydot):
    plt.text(year, ydot, f'{ydot:.2f}', color='grey', ha='right', va='bottom')

# Adding captions and headings
plt.title('Launch success yearly trend', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Success rate', fontsize=20)

# Show polt
plt.grid(True)
plt.tight_layout()
plt.show()
```

You can observe that the success rate since 2013 kept increasing till 2017 (stable in 2014) and after 2015 it started increasing.
